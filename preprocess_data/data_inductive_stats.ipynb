```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("/home/ayush/DyGLib/processed_data/uci/ml_uci.csv")
```

```python
x = df["u"].unique().tolist() + df["i"].unique().tolist()
x = np.unique(x)
```

## First time seen

```python
m = np.ones(len(x))*(-1)
for idx, out in enumerate(df[["u", "i"]].to_numpy()):
  u = out[0]
  i = out[1]
  if m[u-1] == -1:
    m[u-1] = idx
  if m[i-1] == -1:
    m[i-1] = idx
```

```python
plt.hist(m, bins=30, color='skyblue', edgecolor='black')
```

## Last time seen

```python
m = np.ones(len(x))*(-1)
for idx, out in reversed(list(enumerate(df[["u", "i"]].to_numpy()))):
  u = out[0]
  i = out[1]
  if m[u-1] == -1:
    m[u-1] = idx
  if m[i-1] == -1:
    m[i-1] = idx
```

```python
plt.hist(m, bins=30, color='skyblue', edgecolor='black')
```
